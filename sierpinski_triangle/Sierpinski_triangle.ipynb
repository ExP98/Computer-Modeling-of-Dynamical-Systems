```python
'''
1. Take three points in a plane to form a triangle, you need not draw it.
2. Randomly select any point inside the triangle and consider that your current position.
3. Randomly select any one of the three vertex points.
4. Move half the distance from your current position to the selected vertex.
5. Plot the current position.
6. Repeat from step 3.
'''
```

```python
from random import randint
import numpy as np
import matplotlib.pyplot as plt
```

```python
tri_vert = (0.5, 1), (1, 0), (0, 0)
n = int(1e5)
x, y = 0.5, 0.5
```

```python
dots_x, dots_y = [], []
for i in range(n):
    if (i+1) % 10000 == 0:
        print(i+1, end='\r')
    ind = randint(0, 2)
    x, y = (x + tri_vert[ind][0]) / 2, (y + tri_vert[ind][1]) / 2
    dots_x.append(x)
    dots_y.append(y)
```

```python
s = 800
mg = 0.05
xa, xb = min([cd[0] for cd in tri_vert]) - mg, max([cd[0] for cd in tri_vert]) + mg
ya, yb = min([cd[1] for cd in tri_vert]) - mg, max([cd[1] for cd in tri_vert]) + mg
d_x, d_y = xb - xa, yb - ya
print(xa, xb, ya, yb)
matrix = np.zeros((s, s))

for i, j in zip(dots_x, dots_y):
    x_px, y_px = int(s * (i - xa) / d_x), int(- s * (j - ya) / d_y)
    matrix[y_px, x_px] = 1
```

```python
fig, ax = plt.subplots(figsize=(11.69, 8.27))
fig.suptitle(f"Sierpiński triangle, {n} dots")
ax.imshow(matrix, cmap='binary')
ax.set_xticks([])
ax.set_yticks([])
plt.show()
# fig.savefig('output/triangle_'+ str(n) + '.png')
```
